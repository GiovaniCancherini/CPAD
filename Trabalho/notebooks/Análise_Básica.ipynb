```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/prouni_tratado_2017_2020.csv')
```

```python
df["modalidade_ensino_bolsa"].unique()
```

```python
df.columns
```

```python
df['raca_beneficiario_bolsa'].replace('NÃ£o Informada', 'Não informado', inplace=True)
```

```python
df['raca_beneficiario_bolsa'].unique()
```

```python
df['tipo_bolsa'].replace('BOLSA PARCIAL 50%', 'PARCIAL', inplace=True)
df['tipo_bolsa'].replace('BOLSA INTEGRAL', 'INTEGRAL', inplace=True)
```

```python
plt.figure(figsize=(10, 6))
plt.title('Quantidade de Alunos por Modalidade de Ensino', fontsize=16)
df['modalidade_ensino_bolsa'].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.xlabel('Modalidade de Ensino', fontsize=12)
plt.ylabel('Quantidade de Alunos', fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
plt.title('Distribuição de Bolsas por Raça/Cor dos Beneficiários', fontsize=16)
df['raca_beneficiario_bolsa'].value_counts().plot(kind='bar', color=sns.color_palette('pastel'))
plt.xlabel('Raça/Cor Declarada', fontsize=12)
plt.ylabel('Quantidade de Bolsas Concedidas', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
# Agrupar por ano e modalidade de ensino, depois contar o número de alunos
alunos_por_modalidade_ano = df.groupby(['ano_concessao_bolsa.1', 'modalidade_ensino_bolsa']).size().unstack(fill_value=0)

# Assegurar que as colunas existem antes de tentar acessá-las
modalidades_presentes = ['PRESENCIAL', 'EDUCAÇÃO A DISTÂNCIA']
colunas_para_plotar = [m for m in modalidades_presentes if m in alunos_por_modalidade_ano.columns]

if colunas_para_plotar:
    plot_data = alunos_por_modalidade_ano[colunas_para_plotar]
else:
    plot_data = alunos_por_modalidade_ano

# Plotar o gráfico de barras
if not plot_data.empty:
    plot_data.plot(kind='bar', figsize=(14, 8), colormap='viridis')
    plt.title('Evolução Anual de Alunos por Modalidade de Ensino (Presencial vs EAD)', fontsize=16)
    plt.xlabel('Ano de Concessão da Bolsa', fontsize=12)
    plt.ylabel('Quantidade de Alunos', fontsize=12)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(fontsize=10)
    plt.legend(title='Modalidade de Ensino', title_fontsize='13', fontsize='11')
    plt.grid(axis='y', linestyle='--')
    plt.tight_layout()
    plt.show()
else:
    print("Não há dados suficientes para plotar o gráfico de alunos por modalidade e ano.")
```

# Análise de sexo dos beneficiários
Analisando a distribuição de sexo dos beneficiários de bolsas do Prouni.

```python
plt.figure(figsize=(8, 8))
df['sexo_beneficiario_bolsa'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'], wedgeprops=dict(width=0.3), startangle=90, textprops={'fontsize': 12})
plt.title('Distribuição de Bolsas por Sexo dos Beneficiários', fontsize=16)
plt.ylabel('')  # Remove o label do eixo y para tortas
plt.legend(labels=df['sexo_beneficiario_bolsa'].value_counts().index, title="Sexo", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=10, title_fontsize=12)
plt.axis('equal') # Assegura que o gráfico de pizza seja um círculo.
plt.tight_layout()
plt.show()
```

# Análise de tipo de bolsa por raça
Este gráfico mostra a distribuição de tipos de bolsa (Integral ou Parcial) entre as diferentes raças/etnias dos beneficiários.

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='raca_beneficiario_bolsa', hue='tipo_bolsa', palette='muted')
plt.title('Distribuição de Tipos de Bolsa por Raça/Cor dos Beneficiários', fontsize=16)
plt.xlabel('Raça/Cor Declarada', fontsize=12)
plt.ylabel('Quantidade de Bolsas Concedidas', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Tipo de Bolsa', title_fontsize='13', fontsize='11')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

# Análise de modalidade de ensino por sexo
Este gráfico explora como as modalidades de ensino (Presencial e EAD) se distribuem entre os sexos masculino e feminino.

```python
plt.figure(figsize=(10, 7))
sns.countplot(data=df, x='modalidade_ensino_bolsa', hue='sexo_beneficiario_bolsa', palette='pastel')
plt.title('Modalidade de Ensino por Sexo do Beneficiário', fontsize=16)
plt.xlabel('Modalidade de Ensino', fontsize=12)
plt.ylabel('Quantidade de Alunos', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Sexo do Beneficiário', title_fontsize='13', fontsize='11')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```
