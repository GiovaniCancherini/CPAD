```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['raca_beneficiario_bolsa'].replace('NÃ£o Informada', 'Não informado', inplace=True)
df['raca_beneficiario_bolsa'].replace('IndÃ\xadgena', 'Indigena', inplace=True)
```

```python
df['tipo_bolsa'].replace('BOLSA PARCIAL 50%', 'PARCIAL', inplace=True)
df['tipo_bolsa'].replace('BOLSA INTEGRAL', 'INTEGRAL', inplace=True)
```

```python
# Clean the 'modalidade_ensino_bolsa' column by mapping 'educaÃ§Ã£o a distÃ¢ncia' to 'ead'.
# This ensures consistency for this plot and subsequent data analysis.
df['modalidade_ensino_bolsa'].replace('educaÃ§Ã£o a distÃ¢ncia', 'ead', inplace=True)

plt.figure(figsize=(10, 6))

# Generate the pie chart from the 'modalidade_ensino_bolsa' value counts.
# autopct='%1.1f%%' displays the percentage for each slice.
df['modalidade_ensino_bolsa'].value_counts().plot(kind='pie', colormap='viridis', autopct='%1.1f%%', startangle=90)

plt.title('Distribuição de Alunos por Modalidade de Ensino', fontsize=16)
plt.ylabel('')  # Remove the y-axis label as it's not needed for pie charts.
plt.axis('equal')  # Ensures the pie chart is drawn as a circle.
plt.tight_layout() # Adjust plot to ensure everything fits without overlapping.
plt.show()
```

```python
# Gráfico das 5 universidades que mais proporcionaram bolsas com proporção EAD/Presencial
top_5_ies_names = df['nome_ies_bolsa'].value_counts().head(5).index
df_top_5 = df[df['nome_ies_bolsa'].isin(top_5_ies_names)]

# Agrupar por IES e modalidade de ensino
ies_modalidade_counts = df_top_5.groupby(['nome_ies_bolsa', 'modalidade_ensino_bolsa']).size().unstack(fill_value=0)

# Reordenar para manter a ordem das top 5 IES
ies_modalidade_counts = ies_modalidade_counts.reindex(top_5_ies_names)

# Garantir que ambas as modalidades estejam presentes como colunas, mesmo que uma delas tenha contagem zero para alguma IES
# e para manter uma ordem consistente no gráfico (Presencial primeiro)
modalidades_esperadas = ['presencial', 'ead']
for mod in modalidades_esperadas:
    if mod not in ies_modalidade_counts.columns:
        ies_modalidade_counts[mod] = 0

# Selecionar e reordenar colunas para o plot
ies_modalidade_counts = ies_modalidade_counts[modalidades_esperadas]

# Plotar o gráfico de barras agrupado
ax = ies_modalidade_counts.plot(kind='bar', figsize=(15, 10), colormap='viridis')

plt.title('Top 5 Universidades: Bolsas por Modalidade (Presencial vs EAD)', fontsize=18) # Increased title font
plt.xlabel('Universidade', fontsize=14) # Increased label font
plt.ylabel('Número de Bolsas', fontsize=14) # Increased label font
plt.xticks(rotation=45, ha='right', fontsize=11) # Increased tick font
plt.yticks(fontsize=11) # Increased tick font
plt.legend(title='Modalidade de Ensino', title_fontsize='14', fontsize='12') # Increased legend font
plt.grid(axis='y', linestyle='--')

# Adicionar o número em cima de cada barra
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=9, padding=3)

plt.tight_layout()
plt.show()
```

```python
# Gráfico das 5 universidades que mais proporcionaram bolsas
top_5_ies = df['nome_ies_bolsa'].value_counts().head(5)

plt.figure(figsize=(12, 8))
bars = plt.bar(top_5_ies.index, top_5_ies.values, color=sns.color_palette('viridis', 5))
plt.title('Top 5 Universidades por Número de Bolsas Concedidas', fontsize=16)
plt.xlabel('Universidade', fontsize=12)
plt.ylabel('Número de Bolsas', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')

# Adicionar o número em cima de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.05 * top_5_ies.values.max(), int(yval), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Agrupar por ano e modalidade de ensino, depois contar o número de alunos
alunos_por_modalidade_ano = df.groupby(['ano_concessao_bolsa', 'modalidade_ensino_bolsa']).size().unstack(fill_value=0)

# Assegurar que as colunas existem antes de tentar acessá-las
modalidades_presentes = ['PRESENCIAL', 'EDUCAÇÃO A DISTÂNCIA']
colunas_para_plotar = [m for m in modalidades_presentes if m in alunos_por_modalidade_ano.columns]

if colunas_para_plotar:
    plot_data = alunos_por_modalidade_ano[colunas_para_plotar]
else:
    plot_data = alunos_por_modalidade_ano

# Plotar o gráfico de barras
if not plot_data.empty:
    plot_data.plot(kind='bar', figsize=(14, 8), colormap='viridis')
    plt.title('Evolução Anual de Alunos por Modalidade de Ensino (Presencial vs EAD)', fontsize=16)
    plt.xlabel('Ano de Concessão da Bolsa', fontsize=12)
    plt.ylabel('Quantidade de Alunos', fontsize=12)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(fontsize=10)
    plt.legend(title='Modalidade de Ensino', title_fontsize='13', fontsize='11')
    plt.grid(axis='y', linestyle='--')
    plt.tight_layout()
    plt.show()
else:
    print("Não há dados suficientes para plotar o gráfico de alunos por modalidade e ano.")
```

```python
# Calcular a contagem de alunos por ano e modalidade
modalidade_ano_counts = df.groupby(['ano_concessao_bolsa', 'modalidade_ensino_bolsa']).size().unstack(fill_value=0)

# Garantir que ambas as modalidades estejam presentes como colunas e na ordem desejada
modalidades_esperadas = ['presencial', 'ead']
for mod in modalidades_esperadas:
    if mod not in modalidade_ano_counts.columns:
        modalidade_ano_counts[mod] = 0
# Selecionar e reordenar colunas para o plot, garantindo a ordem
modalidade_ano_counts = modalidade_ano_counts[modalidades_esperadas]

# Calcular as proporções percentuais para cada ano (total do ano = 100%)
modalidade_ano_proportions = modalidade_ano_counts.apply(lambda x: (x / x.sum() * 100) if x.sum() > 0 else x, axis=1)

# Plotar o gráfico de barras empilhadas de 100%
ax = modalidade_ano_proportions.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='viridis')

plt.title('Proporção Percentual de Alunos por Modalidade de Ensino (EAD vs Presencial) por Ano', fontsize=16)
plt.xlabel('Ano de Concessão da Bolsa', fontsize=12)
plt.ylabel('Porcentagem de Alunos (%)', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(np.arange(0, 101, 10), [f'{y}%' for y in np.arange(0, 101, 10)], fontsize=10)
plt.legend(title='Modalidade de Ensino', title_fontsize='13', fontsize='11', loc='upper left', bbox_to_anchor=(1.02, 1))
plt.grid(axis='y', linestyle='--')

# Adicionar o rótulo percentual em cada segmento da barra
for c in ax.containers:
    labels = [f'{v.get_height():.1f}%' if v.get_height() > 0.5 else '' for v in c]
    ax.bar_label(c, labels=labels, label_type='center', fontsize=9, color='white', fontweight='bold')

plt.tight_layout(rect=[0, 0, 0.85, 1]) # Ajustar layout para dar espaço para a legenda externa
plt.show()
```
